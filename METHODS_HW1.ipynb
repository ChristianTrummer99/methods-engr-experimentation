```python
# 3 a) b) & c)

t = np.arange(0,30,0.001)
tau = 1.4
omega = 0.7
K = 1
A = 10
T = 330 + A*np.sin(omega*t)
phi = -m.atan(omega*tau)
beta = phi/omega
E = 330 + (-30 + (omega*tau*K*A)/(1 + omega**2*tau**2))*np.exp(-t/tau) + K*A/np.sqrt(1 + omega**2*tau**2)*np.sin(omega*t + phi)# + K*A + (300 - K*A)*np.exp(-t/tau)

# code from lecture: (it's the same as mine)
# y = (y0+(omega*tau*K*A)/(omega**2*tau**2+1))*numpy.exp(-t/tau)+K*A/(numpy.sqrt(omega**2*tau**2+1))*numpy.sin(omega*t+phi)

plt.plot(t,E,c='g',label="E(t) [mV]")
plt.plot(t,T, c='b', label='T(t) [K]')
plt.xlabel('Time [s]')
plt.title('Thermocouple Output')
plt.grid()
plt.gcf().set_size_inches(11,8)
plt.legend(loc="best")
print('a) omega =',omega, 'rad/s')
print('c) beta =',beta,'s\n\nb)')
```

3 d)

The dynamic error, 
\begin{align*}
\epsilon_f(\omega) = 1 - \frac{1}{\sqrt{1+\omega^2 \tau^2}}
\end{align*}
is equivalent to 
\begin{align*}
1 - M(\omega) =1- \frac{y - y_0}{KA}
\end{align*}
and thus we need to use the magnitude ratio form that is time-dependent: 
\begin{align*}
M(t) = \frac{y(t)-y_0}{y_\infty - y_0} = 1 - e^{-t/\tau}
\end{align*}
therefore, 
\begin{align*}
\epsilon_f(t) = 1- \frac{y(t)-y_0}{y_\infty - y_0}
\end{align*}


```python
# 3 d)


dyn_err = 1 - (E - 330)/(K*A)

plt.plot(t,dyn_err,c='r')
plt.grid()
plt.gcf().set_size_inches(11,8)
plt.title('Dynamic Error')
plt.xlabel('Time [s]')
plt.ylabel('Error')
```
