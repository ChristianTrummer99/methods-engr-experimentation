```python
import math as m, numpy as np, matplotlib.pyplot as plt
```

```python
#1 a)

N = 2048 #number of data points
fs = 20 #sampling frq
#we can calculate the sampling interval:
t = np.linspace(0,N/fs,N)#sampling interval
f = 0.5 + 0.6366*np.cos(2*m.pi*t) + 0.1273*np.cos(10*m.pi*t) - 0.0909*np.cos(14*m.pi*t)

sp = np.fft.fft(f) # compute FFT 

k = np.arange(int(N))
frq = k/(N/fs) # two sides frequency range
frq = frq[range(int(N/2))] # one side frequency range
sp1 = sp[range(int(N/2))]


plt.plot(frq,abs(sp1)*2/N,'b.') # plotting the spectrum
plt.xlabel('Freq (Hz)',fontsize='15')
plt.ylabel('|sp(freq)|',fontsize='15');
plt.grid()
plt.gcf().set_size_inches(11,8)
```

1 b) There is no aliasing because aliasing occurs where signal 
frequencies exceed half of the sampling frequency. Noticing the 
frequency of each term in the function, we see that 1, 5, and 7 Hz (the frequencies for each harmonic) don't exceed half of the sampling frequency - 10Hz.



1 c) There is leakage because 2048 is not an integer multiple of 20 [Hz].This means that the total sampling period is not a mulitple of the period in the signal - it does not begin and end at consecutive points. Leakage can be minimized by adjusting the sampling period to adjust the frequency resolution, d) specifically, a new sampling frequency of 40.96 Hz, an integer factor of 2048, creates a nearly perfect spectrum.



2 a) Because the sampling rate is now 4.9 Hz, there is going to be aliasing on the 5th and 7th harmonic because they both exceed half the sampling rate - 2.45 Hz < 5 Hz and 2.45 < 7 Hz.

```python
#2 b) & c)

N = 2048 #number of data points
fs = 4.9 #sampling frq

#we can calculate the sampling interval:
t = np.linspace(0,N/fs,N)#sampling interval
f = 0.5 + 0.6366*np.cos(2*m.pi*t) + 0.1273*np.cos(10*m.pi*t) - 0.0909*np.cos(14*m.pi*t)

t0 = np.linspace(0,N/fs,N**2)#sampling interval
f0 = 0.5 + 0.6366*np.cos(2*m.pi*t0) + 0.1273*np.cos(10*m.pi*t0) - 0.0909*np.cos(14*m.pi*t0)

sp = np.fft.fft(f) # compute FFT 

k = np.arange(int(N))
frq = k/(N/fs) # two sides frequency range
frq = frq[range(int(N/2))] # one side frequency range
sp1 = sp[range(int(N/2))]


#signal
plt.plot(t0[:100*N],f0[:100*N],label="Original")  
plt.scatter(t[:100],f[:100],)
plt.plot(t[:100],f[:100],c='r', label="Sample")
plt.xlim(0,5)
plt.xlabel('t [s]',fontsize='15')
plt.ylabel('f(t)',fontsize='15');
plt.grid()
plt.gcf().set_size_inches(11,8)
plt.legend(loc='best')
plt.show()

#spectrum
plt.plot(frq,abs(sp1)*2/N,'b.') # plotting the spectrum
plt.xlabel('Freq (Hz)',fontsize='15')
plt.ylabel('|sp(freq)|',fontsize='15');
plt.grid()
plt.gcf().set_size_inches(11,8)
```

2 d) the sampled signal, due to very low frequency resolution, is missing the periodicity of the 5th and 7th harmonics. Thus the only recognizeable terms in the sampled signal are: $f\left(t\right)=0.5+0.6366\cos(2\pi t)$ essentially forgetting the $0.1273 \cos(10πt) − 0.0909 \cos(14πt)$ terms.
