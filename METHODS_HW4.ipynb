1. If you are using an oscilloscope while collecting data, you will be able to see if the amplitude or offset is large enough to induce clipping. Typically if the amplitude + offset is larger than 5V, clipping will occur, however, it depends on the device on which you are viewing the wave. More simply,if the original signal lies outside of the range of the A/D converter, it is clipped at the extreme value. Any voltage greater than the upper limit of the A/D converter is assigned the maximum voltage of the A/D converter.
